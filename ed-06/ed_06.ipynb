```python
from typing import Tuple, List
import random
import numpy as np

PESO_MAXIMO: int = 269

def get_ei() -> List[Tuple[int]]:
    return []

def f(inventario: List[Tuple[int]]) -> int:
    (v_total, p_total) = status_inventario(inventario)    
    
    if p_total > PESO_MAXIMO:
        return 0

    return v_total / p_total

def funcao_sucessora(inventario: List[Tuple[int]], items):
    if status_inventario(inventario)[1] <= PESO_MAXIMO:
        inventario.append(items.pop(random.randint(0, len(items)-1)))
    else:
        inventario.append(items.pop(random.randint(0, len(items)-1)))
        items.append(inventario.pop(random.randint(0, len(inventario)-1)))

    return inventario

def status_inventario(inventario: List[Tuple[int]]) -> Tuple[int]:
    v_invent = 0
    p_invent = 0

    for x in inventario:
        v_invent = v_invent + x[0]
        p_invent = p_invent + x[1]

    return (v_invent, p_invent)
```

```python
# Hill climbing
items = [(55, 95), (10,4), (4, 60), (5, 32), (4, 23), (50, 72), (8, 80), (61, 62), (85, 65), (87, 46)]

def hill_climbing():
    iteracoes = 0
    estado_atual = estado_melhor = get_ei()
    while iteracoes < 500:
        vizinhanca = funcao_sucessora(estado_atual, items)
        
        if f(vizinhanca) > f(estado_atual) and f(vizinhanca) > f(estado_melhor):
            estado_melhor = vizinhanca

        v_estado_melhor = f(estado_melhor)
        
        if v_estado_melhor >= PESO_MAXIMO - 10:
            return estado_melhor
        else:
            estado_atual = estado_melhor

        print(f'{iteracoes:03d} - {estado_melhor} - {f(estado_melhor)}')
        iteracoes = iteracoes + 1

hill_climbing()
```

```python
items = [(55, 95), (10,4), (4, 60), (5, 32), (4, 23), (50, 72), (8, 80), (61, 62), (85, 65), (87, 46)]

def simulated_annealing():
    T = Ti = 99999
    fator = 0.7

    estado_melhor = estado_atual = get_ei()

    while T != 0.000001:
        vizinhanca = funcao_sucessora(estado_atual, items)

        if f(vizinhanca) >= f(estado_atual) and f(vizinhanca) >= f(estado_melhor):
            estado_melhor = vizinhanca
        else:
            limite_aceito = np.exp((f(vizinhanca) - f(estado_atual)) / T)
            if random() < limite_aceito:
                estado_melhor = vizinhanca

        if T < 0.000001:
            T = 0.000001
        else:
            T = T * fator

        if f(estado_melhor) >= PESO_MAXIMO - 10:
            return estado_melhor

        print(f'{estado_melhor} - {f(estado_melhor)}')

simulated_annealing()
    
```
